```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy['Age'] = df_copy['Age'].fillna(df_copy['Age'].median())
df_copy['Embarked'] = df_copy['Embarked'].fillna("S")
df_copy['Cabin'] = df_copy['Cabin'].fillna("Unknown")
df_copy["Survived"].replace({1: "Survived", 0: "Not"}, inplace=True)


```

```python
df_copy.loc[ df_copy['Age'] <= 15, 'perAge'] = 15
df_copy.loc[(df_copy['Age'] > 15) & (df_copy['Age'] <= 25), 'perAge'] = 25
df_copy.loc[(df_copy['Age'] > 25) & (df_copy['Age'] <= 35), 'perAge'] = 35
df_copy.loc[(df_copy['Age'] > 35) & (df_copy['Age'] <= 45), 'perAge'] = 45
df_copy.loc[(df_copy['Age'] > 45) & (df_copy['Age'] <= 55), 'perAge'] = 55
df_copy.loc[(df_copy['Age'] > 55) & (df_copy['Age'] <= 65), 'perAge'] = 65
df_copy.loc[(df_copy['Age'] > 65) & (df_copy['Age'] <= 75), 'perAge'] = 75
df_copy.loc[(df_copy['Age'] > 75) & (df_copy['Age'] <= 85), 'perAge'] = 85
df_copy.loc[ df_copy['Age'] > 85, 'perAge']=95
df_copy.head(5)
```

```python
from dash.dependencies import Input,Output,State
from logging import PlaceHolder
```

```python
@app.callback(
    Output('myGraph1', 'figure'),
    Input('Dropdown1', 'value')
)
def update_My_AgeDiv(input_value1):
    filter_1_df = df_copy[df_copy['Survived']==input_value1]
    fig=px.histogram(filter_1_df,x='Age',nbins=40,histnorm='percent',template='simple_white')
    return fig
```

```python
@app.callback(
    Output('myGraph2', 'figure'),
    Input('Dropdown2', 'value')
)
def update_My_AgeDiv(input_value1):
    numOfClass = df.groupby(by=input_value1, as_index=False)['Survived'].sum()
    fig=px.bar(x=numOfClass[numOfClass.columns[0]],y=numOfClass[numOfClass.columns[1]],
               color=numOfClass[input_value1],template='simple_white')
    print(numOfClass)
    return fig
```
